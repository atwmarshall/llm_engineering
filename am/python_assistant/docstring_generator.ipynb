````python
# environment

load_dotenv()
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY', 'your-key-if-not-using-env')
os.environ['ANTHROPIC_API_KEY'] = os.getenv('ANTHROPIC_API_KEY', 'your-key-if-not-using-env')
````

````python
# initialize

openai = OpenAI()
OPENAI_MODEL = "gpt-4o-mini"
````

````python
system_message = "You are an assistant that writes docstrings, typehints and unittests for Python code."
system_message += "Respond only with docstrings conforming to PEP8, type hints conforming to PEP484"
system_message += "Be concise, clear and detail the expected inputs and outputs and their datatypes."
````

````python
def user_prompt_for(python):
    user_prompt = "Enhance the code with docstrings, type hints and write unittests."
    user_prompt += "For each function write a new function that is more efficient at the task if possible. \n\n"
    user_prompt += python
    return user_prompt
````

````python
def messages_for(python):
    return [
        {"role": "system", "content": system_message},
        {"role": "user", "content": user_prompt_for(python)}
    ]
````

````python
# write to a file called optimized.cpp

def write_output(python):
    code = python.replace("```python","").replace("```","")
    with open("documented_and_unittested.py", "w") as f:
        f.write(code)
````

````python
def python_admin_gpt(python):    
    stream = openai.chat.completions.create(model=OPENAI_MODEL, messages=messages_for(python), stream=True)
    reply = ""
    for chunk in stream:
        fragment = chunk.choices[0].delta.content or ""
        reply += fragment
        print(fragment, end='', flush=True)
    write_output(reply)
````

````python
python_admin_gpt(function)
````
